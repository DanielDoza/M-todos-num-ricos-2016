```python
import math
print "Por favor ingrese la masa del paracaidista"
m=float(input())
print "Por favor ingrese el coheficiente de resistencia"
c=float(input())
g=9.8
print "¿Desde que tiempo desea calcular?"
t=float(input())
print "¿Cual es el error aproximado tolerable?"
et=float(input())
print "¿Con que intervalos?"
i=input()
veloApr=0
aprA=0
ea=0
```

```python

cont=0
print repr('TIME').rjust(1),repr('S.ANA').rjust(2),repr('S.APR').rjust(3),repr('ERROR.Apr').rjust(4)

while (ea>et or ea==0):
    aux=(c/m*t)
    veloAn=((m*g)/(c))*(1-(math.exp(-1*aux)))
    if(t==0):
        veloApr=0.0
    else:
        veloApr=veloApr+(g-(c/m)*veloApr)*((t+i)-t)
    
    if(veloApr==0):
        ea=0
        aprA=veloApr
    else:
        ea=(((veloApr)-(aprA))/(veloApr))*(100)
        aprA=veloApr

        
    if(ea>et or ea==0):
        print repr(t).rjust(1),repr('%.4f' %veloAn).rjust(2),repr('%.4f' %veloApr).rjust(3),repr('%.6f'%ea).rjust(4)
        
    t+=i
    cont=ea 
    

    
```
